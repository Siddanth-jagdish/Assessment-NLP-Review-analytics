## Extracting key features from reviews and adding a new column called tags which is created from respected reviews using nouns and adjectives
### Aspect based analysis

```python
import pandas as pd
import numpy as np
import spacy
import string
import nltk
from spacy.lang.en.stop_words import STOP_WORDS
from nltk.corpus import wordnet as wn
```

```python
df1 = pd.read_csv("C:/Users/35387/Downloads/archive (6)/Datafiniti_Amazon_Consumer_Reviews_of_Amazon_Products.csv")
```

```python
df1["sentiment"] = df1["reviews.rating"].apply(lambda x: 0 if x < 4 else 1)
```

```python
df = df1[["name","reviews.text","sentiment"]]
```

```python
df
```

```python
punct = string.punctuation
stopwords = list(STOP_WORDS)
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
def clean_it(sent):
    file = nlp(sent)
    
    list1 = []
    for elem in file:
        if elem.lemma_ != "-PRON-":
            i = elem.lemma_.lower().strip()
        else:
            i = elem.lower_
        list1.append(i)
    str1 = ""
    tokens = []
    for elem in list1:
        if elem not in stopwords and elem not in punct:
            tokens.append(elem)
    for ele in tokens:
        str1 =str1+" " +ele 
        str1 = str1.lstrip()
    return str1
```

```python
clean_it("This kindle is light and easy to use especially at the biggest beach!!!")
```

```python
df["review_clean"] = df["reviews.text"].apply(lambda x: clean_it(x))
```

```python
df
```

```python
aspects = []
for sentence in df["review_clean"]:
    doc = nlp(sentence)
    descriptive_term = ''
    target = ''
    for token in doc:
        if token.dep_ == 'nsubj' and token.pos_ == 'NOUN':
            target = token.text
        if token.pos_ == 'ADJ':
            prepend = ''
            for child in token.children:
                if child.pos_ != 'ADV':
                    continue
                prepend += child.text + ' '
            descriptive_term = prepend + token.text
    if target != "" and descriptive_term != "":
        aspects.append({"aspect" : target ,"description" : descriptive_term })
    else:
        aspects.append({"aspect": "0000", "description" : "11111"})
print(aspects)
```

```python
k = aspects
```

```python
nouns = {x.name().split('.', 1)[0] for x in wn.all_synsets('n')}
```

```python
nouns
```

```python
tag_list = []   
```

```python
for i in k:
    if i['aspect'] in nouns:                                               #check if aspect is a noun
        doc = nlp(i['description'])                                        #perform nlp for description if previous line is true
        for word in doc:                                                   #go to each word in description
            flag = 0
            if word.pos_ == 'ADJ':                                         #check if it is an adjective
                tag_list.append(str(word)+" "+i['aspect'])                 #append adjective combined with noun to the tag_list
                flag = 1
                break
        if flag == 0:                                                       #if no adjective found, append "NO ADJECTIVE"
            tag_list.append("NO ADJECTIVE")
    else:
        tag_list.append("TAG UNKNOWN")                                      #if not noun, append "TAG UNKNOWN"
```

```python
tag_list
```

```python
df['Tags'] = tag_list   
```
